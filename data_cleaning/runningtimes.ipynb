```python
import pandas  as pd
import numpy as np
running_times=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/running_times.csv', sep=',', error_bad_lines=False)
running_times.rename(columns={'ReleaseCountry': 'CountryName','RunningTime': 'runtime'}, inplace=True)
running_times.dropna(subset=['runtime'], inplace=True)
running_times[['CountryName']] = running_times[['CountryName']].fillna(value='Other')
running_times.dropna(subset=['ClipId'], inplace=True)
running_times
```

```python
mechok = running_times[running_times['runtime'].astype(str).str.isnumeric() == False].copy()
ok = running_times[running_times['runtime'].astype(str).str.isnumeric() == True].copy()

mechok['runtime']=mechok['runtime'].str.extract('(\d+)(\.|\'|\s*minutes|\s*mins|\s*min)',expand=True)
mechok.dropna(subset=['runtime'],inplace=True)

running_times = pd.concat([ok,mechok])
```

```python
running_times['runtime']=running_times['runtime'].astype(int)
running_times['rid']=(running_times.runtime).rank(method='dense')
running_times['rid']=running_times['rid'].astype(int)
Runningtimes = running_times[['rid','runtime']].copy()
Runningtimes.drop_duplicates(subset=['rid'],keep='first', inplace=True)
Runningtimes.to_csv('Runningtimes.csv', sep=',', encoding='utf-8', index=False)
```

```python
running_times.drop(['runtime'],axis=1,inplace=True)
countries=pd.read_csv('/Users/macbook/Desktop/BA6/Database/newfiles/countries.csv', sep=',', error_bad_lines=False)
running_times
```

```python
running_times=pd.merge(running_times, countries, how='left', on='CountryName')
running_times['countid'] = running_times['countid'].fillna(value='223')
running_times.drop(['CountryName'],axis=1,inplace=True)
clips=pd.read_csv('/Users/macbook/Desktop/BA6/Database/newfiles/clips.csv', sep=',', error_bad_lines=False)
clips.rename(columns={'ClipIds': 'ClipId'}, inplace=True)
running_times=pd.merge(running_times, clips, how='left', on='ClipId')
running_times
```

```python
running_times.dropna(subset=['Title','ClipId'], inplace=True)
running_times.drop(['Title','Year','Type','Votes','Rank'],axis=1,inplace=True)
running_times.drop_duplicates(subset=['rid','ClipId','countid'],keep='first', inplace=True)
running_times.to_csv('runs_for.csv', sep=',', encoding='utf-8', index=False)
```
