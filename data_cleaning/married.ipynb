```python
import pandas  as pd
import numpy as np
```

```python
married=pd.read_csv('/Users/macbook/Desktop/BA6/Database/newfiles/married.csv', sep=',', error_bad_lines=False)
len(married)
married
```

```python
married['from']=married['name'].str.extract("(\d{4}) \-", expand=True)
married['to']=married['name'].str.extract("(\d{4})\)",expand=True)
married['children']=married['name'].str.extract("\;(.*)",expand=True)
married['status']=married['name'].str.extract("\) \((.*)\)",expand=True)
married['name']=married['name'].str.extract("\'(.*)\'",expand=True)
```

```python
married.to_csv('married.csv', sep=',', encoding='utf-8', index=False)
```
