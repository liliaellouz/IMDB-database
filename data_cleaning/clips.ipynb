```python
import pandas  as pd
import numpy as np
```

```python
clips=pd.read_csv('/Users/macbook/Desktop/BA6/Database/newfiles/clips.csv', sep=',', error_bad_lines=False)
ratings=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/ratings.csv', sep=',', error_bad_lines=False)
len(clips)
```

```python
clips= pd.merge(clips, ratings, how='left', on='ClipId')
clips.rename(columns={'ClipId': 'ClipId','ClipTitle': 'Title','ClipYear':'Year','ClipType':'Type'}, inplace=True)
```

```python
clips = clips.replace('', np.nan, regex=True)
clips.drop_duplicates(keep='first', inplace=True)
clips.drop_duplicates(subset=['ClipId'], keep='first', inplace=True)
clips['Year'] = clips['Year'].astype(str).str.extract('(\d{4})')

df = clips[clips['Year'].astype(str).str.isnumeric() == False]
df
```

```python
clips.to_csv('clips.csv', sep=',', encoding='utf-8', index=False)
```

```python
clip_links=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/clip_links.csv', sep=',', error_bad_lines=False)
clip_links
```

```python
clip_links.dropna(subset=['LinkType','ClipFrom','ClipTo'], inplace=True)
clip_links['LinkId']=(clip_links.LinkType).rank(method='dense')
link_types=clip_links[['LinkType','LinkId']].copy()
link_types.drop_duplicates(subset=['LinkId'], keep='first', inplace=True)
link_types=link_types[['LinkId', 'LinkType']]
link_types.to_csv('Link_types.csv', sep=',', encoding='utf-8', index=False)
```

```python
clip_links.drop(['LinkType'],axis=1, inplace=True)
clip_links.drop_duplicates(keep='first', inplace=True)
clip_links.to_csv('Clip_Links.csv', sep=',', encoding='utf-8', index=False)
```

```python
genres=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/genres.csv', sep=',', error_bad_lines=False)
genres['gid']=(genres.Genre).rank(method='dense')
Clip_genre=genres[['gid','Genre']].copy()
Clip_genre.drop_duplicates(subset=['gid'], keep='first', inplace=True)
Clip_genre.to_csv('Clip_genre.csv', sep=',', encoding='utf-8', index=False)
```

```python
genres.drop(['Genre'],axis=1, inplace=True)
genres.drop_duplicates(keep='first', inplace=True)
genres.to_csv('has_genre.csv', sep=',', encoding='utf-8', index=False)
```

```python
filmed_in=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/countries.csv', sep=',', error_bad_lines=False)
filmed_in['countid']=(filmed_in.CountryName).rank(method='dense')
countries=filmed_in[['countid','CountryName']].copy()
countries.drop_duplicates(subset=['countid'], keep='first', inplace=True)
countries.to_csv('countries.csv', sep=',', encoding='utf-8', index=False)
```

```python
filmed_in.drop(['CountryName'],axis=1, inplace=True)
filmed_in.drop_duplicates(keep='first', inplace=True)
filmed_in.to_csv('filmed_in.csv', sep=',', encoding='utf-8', index=False)
```

```python
released_in=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/release_dates.csv', sep=',', error_bad_lines=False)
released_in.rename(columns={'ReleaseCountry': 'CountryName'}, inplace=True)
released_in.dropna(subset=['CountryName'],inplace=True)
countries=pd.read_csv('/Users/macbook/Desktop/BA6/Database/newfiles/countries.csv', sep=',', error_bad_lines=False)
countriestoadd = released_in[np.isnan(released_in['countid'])].copy()
countriestoadd['countid']=(countriestoadd.CountryName).rank(method='dense')+countries['countid'].max()
countriestoadd[['countid']] = countriestoadd[['countid']].astype(int)
countriestoadd= countriestoadd[['countid','CountryName']]
countries = pd.concat([countries, countriestoadd])
countries.drop_duplicates(keep='first', inplace=True)
countries.to_csv('countries.csv', sep=',', encoding='utf-8', index=False)

```
