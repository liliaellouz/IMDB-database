```python
import pandas  as pd
import numpy as np
```

```python
languages=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/languages.csv', sep=',', error_bad_lines=False)
languages[['Language']] = languages[['Language']].replace({'\(':'','\)':''},regex=True)
languages.dropna(subset=['Language'], inplace=True)
languages['lid']=(languages.Language).rank(method='dense')
Clip_languages=languages[['lid','Language']].copy()
Clip_languages.drop_duplicates(subset=['lid'], keep='first', inplace=True)
Clip_languages.dropna(subset=['lid'], inplace=True)
Clip_languages.to_csv('Clip_languages.csv', sep=',', encoding='utf-8', index=False)
languages.drop(['Language'],axis=1, inplace=True)
languages.drop_duplicates(keep='first', inplace=True)
languages.to_csv('has_language.csv', sep=',', encoding='utf-8', index=False)
```
