```python
import pandas  as pd
import numpy as np
```

```python
actors=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/actors.csv', sep=',')

cols_to_check = ['ClipIds','Chars', 'OrdersCredit' , 'AddInfos']
actors[cols_to_check] = actors[cols_to_check].replace({'\[':'','\]':''},regex=True)

actors = actors.replace('', np.nan, regex=True)

actors.drop_duplicates(keep='first', inplace=True)
actors.drop_duplicates(subset=['FullName'], keep='first', inplace=True)

actors['pid']=(actors.FullName).rank(method='dense')
actors[['pid']] = actors[['pid']].astype(int)

people = actors[['pid','FullName']]
```

```python
people.to_csv('people.csv', sep=',', encoding='utf-8', index=False)
```

```python
act = actors.drop(['FullName'],axis=1)

act[['ClipIds']] = act[['ClipIds']].astype(str)
act[['AddInfos']] = act[['AddInfos']].astype(str)
act[['Chars']] = act[['Chars']].astype(str)
act[['OrdersCredit']] = act[['OrdersCredit']].astype(str)

def splitDataFrameList(df,target_column,separator):
    ''' df = dataframe to split,
    target_column = the column containing the values to split
    separator = the symbol used to perform the split
    returns: a dataframe with each entry for the target column separated, with each element moved into a new row. 
    The values in the other columns are duplicated across the newly divided rows.
    '''
    def splitListToRows(row,row_accumulator,target_column,separator):
        split_row = []
        i=0
        for c in target_column:
            split_row.append(row[c].split(separator))
            i+=1
            
        for s in range (0,len(split_row[0])):
            j=0
            new_row = row.to_dict()
            for c in target_column:
                new_row[c] = split_row[j][s]
                j+=1
            row_accumulator.append(new_row)
    new_rows = []
    df.apply(splitListToRows,axis=1,args = (new_rows,target_column,separator))
    global new_df
    new_df = pd.DataFrame(new_rows)
    return new_df

act = splitDataFrameList( act, ['ClipIds','AddInfos','Chars','OrdersCredit'] , '|')
```

```python
act['charid']=(act.Chars).rank(method='dense')
act[['charid']] = act[['charid']].astype(int)

act = act.replace('', np.nan, regex=True)
act = act.replace('nan', np.nan, regex=True)

act = act[['pid', 'ClipIds', 'charid', 'Chars', 'OrdersCredit', 'AddInfos']]

act.drop_duplicates(subset=['pid', 'charid', 'ClipIds'], keep='first')
```

```python
act.to_csv('act.csv', sep=',', encoding='utf-8', index=False)
```
