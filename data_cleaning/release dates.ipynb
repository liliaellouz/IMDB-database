```python
released_in=pd.read_csv('/Users/macbook/Desktop/BA6/Database/db2018imdb/release_dates.csv', sep=',', error_bad_lines=False)
released_in.rename(columns={'ReleaseCountry': 'CountryName'}, inplace=True)
released_in[['CountryName']] = released_in[['CountryName']].fillna(value='Other')
released_in.dropna(subset=['ReleaseDate','ClipId'], inplace=True)

countries=pd.read_csv('/Users/macbook/Desktop/BA6/Database/newfiles/countries.csv', sep=',', error_bad_lines=False)
released_in= pd.merge(released_in, countries, how='left', on='CountryName')
released_in[['countid']]=released_in[['countid']].fillna(value = 223)
released_in.drop(['CountryName'],axis=1,inplace=True)
released_in['ReleaseDate'] = released_in['ReleaseDate'].str.extract('(\d{4})')
released_in['ReleaseDate']=released_in['ReleaseDate'].astype(int)
released_in['did']=(released_in.ReleaseDate).rank(method='dense')
```

```python
Releasedates = released_in[['did','ReleaseDate']].copy()
Releasedates.drop_duplicates(keep='first', inplace=True)
Releasedates.to_csv('Releasedates.csv', sep=',', encoding='utf-8', index=False)
```
